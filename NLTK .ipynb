```python
#Remove numbers
def text_removenumber(text):
    text=re.sub(r'\d+','', text)
    return text
input_text='"There are 6 balls in this bag, and 22 in the other one."'
```

```python
text_removenumber(input_text)
```

```python
#We can also convert the numbers into words.
# import the inflect library 
import inflect 
p=inflect.engine()
# convert number into words 
def convert_number(text):
    temp_str =text.split()                  # split string into list of words 
    new_str=[]                                 # initialise empty list 
    
    
    for word in temp_str:
        if word.isdigit():
            temp=p.number_to_words(word)
            new_str.append(temp)
        # # append the word as it is    
        else:
            new_str.append(word)
            
            # join the words of new_string to form a string 
    temp_str=' '.join(new_str)
    return new_str
        
input_text='There are 44 balls in this bag, and 99 in the other one.' 
convert_number(input_text) 
```

```python
#Remove punctuation:
def remove_punct(text):
    result=str.maketrans('','', string.punctuation)
    return text.translate(result)
input_str="Hey, did you know that the summer break is coming? Amazing right !! It's only 5 more days !!"
remove_punct(input_str)
```

```python
#Remove whitespaces:
def remove_whitespace(text):
    return " ".join(text.split())

input_str = "   we don't need   the given questions"
remove_whitespace(input_str) 
 
```

```python
#Remove default stopwords:
import nltk
nltk.download('all')
from nltk.courpus import stopwords
from nltk.tokenize import word_tokenize

def remove_stopwords(text):
    stop_words=set(stopwords.stop('english'))
    word_tokens= word_tokenize(text)
    filtered_text=[word for word in word_tokens if word not in stop_words]
    return filtered_text

text=' “This is a sample sentence and we are going to remove the stopwords from this”'    
remove_stopwords(text)    
    
```
